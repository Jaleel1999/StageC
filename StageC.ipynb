```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from xgboost import XGBClassifier
import numpy as np
```

```python
df = pd.read_csv("Data_for_UCI_named.csv")
```

```python
df
```

```python
df.head()
```

```python
features = df.drop(columns=["stab","stabf"])
```

```python
features
```

```python
target = df["stabf"]
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
target = le.fit_transform(target)
```

```python
target
```

```python
#Splitting the dataset into a trainingset and testset.
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(features,target,test_size=0.2, random_state=1)
```

```python
X_train
```

```python
#implementing XGBOOSTClassifier.
xgb = XGBClassifier()
```

```python
xgb.fit(X_train,y_train)
```

```python
from sklearn.metrics import recall_score,accuracy_score,precision_score,f1_score,confusion_matrix
```

```python
ypred = xgb.predict(X_test)
```

```python
accuracy = accuracy_score(y_true=y_test,y_pred=ypred)
```

```python
#Accuracy for XGB model.
accuracy
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier()
classifier.fit(X_train, y_train)

# Predicting a new result
ypred1 = classifier.predict(X_test)
```

```python
ypred1
```

```python
accuracy1 = accuracy_score(y_true=y_test,y_pred=ypred1)
```

```python
#Accuracy for RandomForest Classifier.
accuracy1
```

```python
import lightgbm as lgb
clf = lgb.LGBMClassifier()
clf.fit(X_train, y_train)
```

```python
ypred2 = clf.predict(X_test)
```

```python
accuracy2 = accuracy_score(y_true=y_test,y_pred=ypred2)
```

```python
#Accuracy LGBM classifier.
accuracy2
```

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.model_selection import RandomizedSearchCV
```

```python
#Hyperparameter 1 caclulation.
clf2 = ExtraTreesClassifier(n_estimators = 300 , min_samples_split = 5 , min_samples_leaf = 6, max_features = 'auto')
clf2.fit(X_train, y_train)
ypred3 = clf2.predict(X_test)
```

```python
accuracy3 = accuracy_score(y_true=y_test,y_pred=ypred3)
```

```python
accuracy3
```

```python
# define search space
space = dict()

# define search
search = RandomizedSearchCV(clf2,space,cv=5, n_iter=10, scoring = 'accuracy', n_jobs = -1, verbose = 1,random_state=1)
```

```python
search.fit(X_train,y_train)
```

```python
ysc = search.predict(X_test)
```

```python
accuracy4 = accuracy_score(y_true=y_test,y_pred=ysc)
```

```python
accuracy4
```
